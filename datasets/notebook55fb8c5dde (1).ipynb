```python
from ollama import chat
from ollama import ChatResponse
import time
import tiktoken
import psutil
import os

sys_prompt = ""
def generate_content(msg, system_msg="", model_name='qwen3:1.7b'):
    messages = []
    if system_msg != "":
        messages.append({"role": "system", "content": system_msg})
    messages.append({"role": "user", "content": msg})
    
    # Track metrics
    start_time = time.time()
    process = psutil.Process(os.getpid())
    memory_before = process.memory_info().rss / (1024 * 1024)  # MB
    
    # Use tiktoken for accurate token counting
    try:
        encoding = tiktoken.encoding_for_model("gpt-3.5-turbo")  # Default encoding
        
        # Count input tokens accurately
        input_text = msg + system_msg
        input_tokens = len(encoding.encode(input_text))
    except Exception as e:
        # Fall back to estimation if tiktoken fails
        input_tokens = len(msg + system_msg) // 4
        print(f"Tiktoken error (using estimate): {e}")
    
    response: ChatResponse = chat(model=model_name, messages=messages, stream=True)
    res = ""
    output_tokens = 0
    first_token_time = None
    token_timestamps = []
    
    for chunk in response:
        if not first_token_time and chunk.message.content:
            first_token_time = time.time()
            
        content = chunk.message.content
        res += content
        print(content, end='', flush=True)
        
        # Record timestamp for token rate analysis
        if content:
            token_timestamps.append(time.time())
            
        # Track output tokens from each chunk if available
        if hasattr(chunk, 'done') and chunk.done:
            # Some Ollama models return token metrics in the final chunk
            output_tokens = getattr(chunk, 'eval_count', 0) or getattr(chunk, 'total_tokens', 0)
    
    # If we couldn't get token count from API, estimate with tiktoken
    if output_tokens == 0:
        try:
            output_tokens = len(encoding.encode(res))
        except Exception:
            # Fall back to rough estimate
            output_tokens = len(res) // 4
    
    # Calculate metrics
    end_time = time.time()
    memory_after = process.memory_info().rss / (1024 * 1024)  # MB
    memory_used = memory_after - memory_before
    total_latency = end_time - start_time
    first_token_latency = (first_token_time - start_time) if first_token_time else 0
    tokens_per_second = output_tokens / total_latency if total_latency > 0 else 0
    
    # Calculate token generation stability (standard deviation of inter-token times)
    token_intervals = []
    if len(token_timestamps) > 1:
        for i in range(1, len(token_timestamps)):
            token_intervals.append(token_timestamps[i] - token_timestamps[i-1])
    metadata = {
        "input_tokens": input_tokens,
        "output_tokens": output_tokens,
        "total_tokens": input_tokens + output_tokens,
        "memory_used": memory_used,
        "latency": total_latency,
        "first_token_latency": first_token_latency,
        "tokens_per_second": tokens_per_second,
        "token_intervals": token_intervals
    }

    
    return res, metadata



```

```python
import regex
import pandas as pd


data = pd.read_csv("/kaggle/input/sat-maths/sat-maths.csv")
## without any few shot examples
system_msg = "you are a mathematical expert , you are here to solve complex problems accurately , you should think step by step and provide clear explanation and solution for each problem. give final answer in \n answer {...} \n explanation {....} format"
def process_row(row):
    # Extract the relevant columns from the row
    ques = row["question"]
    options = row["options"]
    if options is not None:
        query = f"solve this question {ques} \n given options\n {options}"
    else:
        query = f"solve this question {ques}"

    res , metadata = generate_content(query, system_msg=system_msg, model_name=model_name)
    row["phi4:3.7B_response"] = res
    row["phi4:3.7B_metadata"] = metadata
    try:
        final_answer = regex.findall(r"answer\s*\{(.*?)\}", res)[0]
        row["final_answer_phi4"] = final_answer
        print("Final answer found:", final_answer)
    except IndexError:
        row["final_answer_phi4"] = "No answer found"
        print("No final answer found in the response.")
    return row



```

```python
import subprocess
process = subprocess.Popen("ollama serve", shell=True) #runs on a different thread
#Download model

import ollama
model_name = "phi4-mini-reasoning"
#model_name = "qwen3:1.7b"
data = data.apply(process_row, axis=1)
data.to_csv("/kaggle/input/sat-maths/sat-maths.csv", index=False)
print("Data processing complete. Responses saved to data_phi4.csv")
```
